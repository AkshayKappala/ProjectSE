```python
import cv2
import pywt
import numpy as np
import matplotlib.pyplot as plt

image_file = 'landscape.png'
img = cv2.imread(f'data/images/{image_file}', cv2.IMREAD_UNCHANGED)

dwt_lvl_1_coefficients = pywt.dwt2(img, 'bior2.2')
LL, (LH, HL, HH) = dwt_lvl_1_coefficients

fig_rows = 2
fig_columns = 2
fig = plt.figure(figsize = (10, 10))
fig.add_subplot(fig_rows,fig_columns,1)
plt.imshow(LL)
fig.add_subplot(fig_rows,fig_columns,2)
plt.imshow(LH)
fig.add_subplot(fig_rows,fig_columns,3)
plt.imshow(HL)
fig.add_subplot(fig_rows,fig_columns,4)
plt.imshow(HH)

dwt_lvl_2_coefficients = pywt.dwt2(LL, 'bior2.2')
LL2, (LH2, HL2, HH2) = dwt_lvl_2_coefficients


fig2 = plt.figure(figsize = (10, 10))
fig2.add_subplot(fig_rows,fig_columns,1)
plt.imshow(LL2)
print(len(LL2))
print(type(LL2))
print(LL2.shape)
# for i in range (LL2.shape[2]):
#     for j in range (LL2.shape[0]):
#         for k in range (LL2.shape[1]):
#             LL2[j,k,i] = 0
#             
fig2.add_subplot(fig_rows, fig_columns, 2)
plt.imshow(LH2)
fig2.add_subplot(fig_rows, fig_columns, 3)
plt.imshow(HL2)
fig2.add_subplot(fig_rows, fig_columns, 4)
plt.imshow(HH2)

dwt_lvl_2_coefficients_recon = LL2, (LH2, HL2, HH2)
LL1_recon = pywt.idwt2(dwt_lvl_2_coefficients_recon, 'bior2.2')
dwt_lvl_1_coefficients_recon = LL1_recon, (LH, HL, HH)
reconstructed_img_lvl2 = pywt.idwt2(dwt_lvl_1_coefficients_recon, 'bior2.2')
reconstructed_img_lvl2 = np.uint8(reconstructed_img_lvl2)
cv2.imwrite(f'data/images/tmp/reconstructed_lvl2_{image_file}', reconstructed_img_lvl2)


# Reconstruct the image from the coefficients
reconstructed_img = pywt.idwt2(dwt_lvl_1_coefficients, 'bior2.2')

# Convert the reconstructed image back to uint8 for display and saving
reconstructed_img = np.uint8(reconstructed_img)


# Save the original and reconstructed images
cv2.imwrite(f'data/images/tmp/original_{image_file}', img)
cv2.imwrite(f'data/images/tmp/reconstructed_{image_file}', reconstructed_img)

```
